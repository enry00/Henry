```python
#importer pandas, numpy,matplotlib et stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
#lire le fichier csv du projet
Emplacement_fichier= r'C:\Users\HP\Desktop\Projet info eivp1\EIVP_KM.csv'
Em_f=Emplacement_fichier
df= pd.read_csv(Em_f,sep=";",index_col='sent_at',parse_dates=True)
df
```

```python
#Afficher les valeurs statistiques
df[['id','noise','temp','humidity','lum','co2',]].describe()
```

```python
# Afficher la courbe (tempÃ©rature en fonction de la date et les variables statistiques)
vs=df['2020-09-11':'2020-09-25']['temp'].resample('w').agg(['mean','max','min','std'])
plt.figure(figsize=(15,5))
df['2020-09-11':'2020-09-25']['temp'].plot(label='temperature')
plt.fill_between(vs.index,vs['max'],vs['min'],alpha=0.2,label='minimum-maximum')
df['2020-09-11':'2020-09-25']['temp'].resample('w').std().plot(alpha=0.8,label='Ã©cart type')
df['2020-09-11':'2020-09-25']['temp'].resample('w').mean().plot(alpha=2,label='moyenne')
#   start date : end date
plt.legend()
plt.show()
df['2020-09-11':'2020-09-25']['temp'].describe()
```

```python
# Afficher la courbe (noise en fonction de la date et les variables statistiques)
vs=df['2020-09-11':'2020-09-25']['noise'].resample('w').agg(['mean','max','min','std'])
plt.figure(figsize=(15,5))
df['2020-09-11':'2020-09-25']['noise'].plot(label='noise')
plt.fill_between(vs.index,vs['max'],vs['min'],alpha=0.2,label='minimum-maximum')
df['2020-09-11':'2020-09-25']['noise'].resample('w').std().plot(alpha=0.8,label='Ã©cart type')
df['2020-09-11':'2020-09-25']['noise'].resample('w').mean().plot(alpha=2,label='moyenne')
#   start date : end date
plt.legend()
plt.show()
df['2020-09-11':'2020-09-25']['noise'].describe()
```

```python
# Afficher la courbe (humiditÃ© en fonction de la date et les variables statistiques)
vs=df['2020-09-11':'2020-09-25']['humidity'].resample('w').agg(['mean','max','min','std'])
plt.figure(figsize=(15,10))
df['2020-09-11':'2020-09-25']['humidity'].plot(label='humidity')
plt.fill_between(vs.index,vs['max'],vs['min'],alpha=0.2,label='minimum-maximum')
df['2020-09-11':'2020-09-25']['humidity'].resample('w').std().plot(alpha=0.8,label='Ã©cart type')
df['2020-09-11':'2020-09-25']['humidity'].resample('w').mean().plot(alpha=2,label='moyenne')
#   start date : end date
plt.legend()
plt.show()
df['2020-09-11':'2020-09-25']['humidity'].describe()
```

```python
# Afficher la courbe (lum en fonction de la date et les variables statistiques)
vs=df['2020-09-11':'2020-09-25']['lum'].resample('w').agg(['mean','max','min','std'])
plt.figure(figsize=(15,5))
df['2020-09-11':'2020-09-25']['lum'].plot(label='humidity')
plt.fill_between(vs.index,vs['max'],vs['min'],alpha=0.2,label='minimum-maximum')
df['2020-09-11':'2020-09-25']['lum'].resample('w').std().plot(alpha=0.8,label='Ã©cart type')
df['2020-09-11':'2020-09-25']['lum'].resample('w').mean().plot(alpha=2,label='moyenne')
#   start date : end date
plt.legend()
plt.show()
df['2020-09-11':'2020-09-25']['lum'].describe()
```

```python
# Afficher la courbe (co2 en fonction de la date et les variables statistiques)
vs=df['2020-09-11':'2020-09-25']['co2'].resample('w').agg(['mean','max','min','std'])
plt.figure(figsize=(15,10))
df['2020-09-11':'2020-09-25']['co2'].plot(alpha=0.7,label='humidity')
plt.fill_between(vs.index,vs['max'],vs['min'],alpha=0.2,label='minimum-maximum')
df['2020-09-11':'2020-09-25']['co2'].resample('w').std().plot(alpha=0.8,label='Ã©cart type')
df['2020-09-11':'2020-09-25']['co2'].resample('w').mean().plot(alpha=2,label='moyenne')
#   start date : end date
plt.legend()
plt.show()
df['2020-09-11':'2020-09-25']['co2'].describe()
```
